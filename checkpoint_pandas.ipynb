```python
import numpy as np
import pandas as pd
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
            'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
            'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
            'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels =    ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = df = pd.DataFrame(exam_data)
df['labels']=labels
df = df.set_index('labels')

#question 1
print(df.head(3))

#question 2
print(df.dropna(subset=['score']))

#question 3
print(df[['name','score']])

#question 4
dict = {'name' : "Suresh", 'score': 15.5, 'attempts': 1, 'qualify': "yes"}
row = pd.DataFrame([dict],index = ['k'])
print(pd.concat([df , row]))

#question 5
print(df.drop('attempts',axis = 1))

#question 6
success_list = []
for i in range (len(exam_data['score'])) :
    if exam_data['score'][i] > 10 :
        success_list.append('1')
    elif exam_data['score'][i] < 10 or  exam_data['score'][i]==10:
        success_list.append('0')
    else:
        success_list.append('empty')
df['success'] = success_list
print(df)

#question 7
df.to_csv(r'C:\Users\am\Desktop\Data.csv')
new_data=pd.read_csv(r'C:\Users\am\Desktop\Data.csv')
print(new_data)






```
